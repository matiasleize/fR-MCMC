Here we are going to compare the integration for the Hu-Sawicki model (n=1) with different initial conditions.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import interp1d

import os
import git
path_git = git.Repo('.', search_parent_directories=True).working_tree_dir
path_datos_global = os.path.dirname(path_git)

os.chdir(path_git); os.sys.path.append('./Software/utils/')

from int_sist_1 import integrador
```

Let's plot the solution of the numerical integrarion for different values of the initial redshift zi.

```python
#%% Hu-Sawicki (n=1)
params_fisicos = [0.3, 0.1, 73] # [omega_m, b, H0]
zs_ode, H_HS = integrador(params_fisicos, verbose=True, model='HS', z_inicial=10)
zs_ode_1, H_HS_1 = integrador(params_fisicos, verbose=True, model='HS', z_inicial=20)
zs_ode_2, H_HS_2 = integrador(params_fisicos, verbose=True, model='HS', z_inicial=30)
#%% Graficamos todos los datos juntos
#%matplotlib qt5
plt.figure()
plt.title('Integrador $f(R)$')
plt.xlabel('z (redshift)')
plt.ylabel('H(z) ((km/seg)/Mpc)')



plt.plot(zs_ode_2,H_HS_2,'.',label='HS=30')
plt.plot(zs_ode_1,H_HS_1,'.',label='HS=20')
plt.plot(zs_ode,H_HS,'.',label='HS=10')
plt.legend(loc = 'best')
plt.grid(True)
#plt.show()
```

The relative error is calculated as follows:

```python
plt.figure()
f_30 = interp1d(zs_ode_2, H_HS_2)
f_20 = interp1d(zs_ode_1, H_HS_1)


plt.plot(zs_ode,1-H_HS/f_30(zs_ode),'.',label='HS=10/30')
plt.plot(zs_ode,1-H_HS/f_20(zs_ode),'.',label='HS=10/20')
plt.legend(loc = 'best')
plt.grid(True)
plt.show()
```
